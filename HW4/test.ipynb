```python
import math

def norm_pdf(x, mean, sd):
    return (1/(math.sqrt(sd*2*math.pi)))*(math.exp(-(((x-mean)**2)/(2*sd))))
```

```python
# maybe coming up with planets around each star
mean = 3
sd = .75

vals = [np.random.normal(mean, sd) for _ in range(1000)]
vals   # need to convert into discrete variable, bin range of values
# count how many random values fall within each bin
# bin becomes outcome, prob is how many values generated within each bin
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.displot(vals, bins=100)
plt.show()
```

```python
# need to convert value into drv, need to define bins
# use linspace to define bins
minv = min(vals)
maxv = max(vals)

np.linspace(minv, maxv, 10)
# 1000 values, 10 outcomes, have to decide which bin random numbers go into
# find value in sequence that is closet to which ever random value
# keep track of counts
```

```python
# trick for finding closest
bins = np.linspace(minv, maxv, 10)
prob = defaultdict(lambda: 0)
for val in vals:
    bin = min(bins, key=lambda x: abs(x - val))
    prob[bin] += 1
    

prob = {key: prob[key] / sum(prob.values()) for key in prob.keys()}
prob
```

```python
import random
dist = {0: 0.20, 2: 0.50, 5: 0.30}
import seaborn as sns
def rand_sample():
    return random.choices(population=list(dist.keys()), weights=list(dist.values()), k=1)[0]

sample = [rand_sample() for i in range(20)]
    
sns.displot(sample, kind='hist', stat='probability', bins=20)
```
